### Step 2: Causal Analysis with DoWhy
We'll define a causal graph to model relationships and estimate the effect of `Duration`, `Heart_Rate`, and `Body_Temp` on `Calories`, adjusting for confounders like `Age`, `Sex`, `Height`, and `Weight`.

```python
import pandas as pd
import numpy as np
from dowhy import CausalModel
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv("train.csv")

# Convert 'Sex' to binary (male=1, female=0)
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})

# Define the causal graph
graph = """
digraph {
    Age;
    Sex;
    Height;
    Weight;
    Duration;
    Heart_Rate;
    Body_Temp;
    Calories;
    Age -> Duration;
    Age -> Heart_Rate;
    Age -> Body_Temp;
    Sex -> Duration;
    Sex -> Heart_Rate;
    Sex -> Body_Temp;
    Height -> Duration;
    Height -> Heart_Rate;
    Weight -> Duration;
    Weight -> Heart_Rate;
    Duration -> Heart_Rate;
    Duration -> Body_Temp;
    Duration -> Calories;
    Heart_Rate -> Body_Temp;
    Heart_Rate -> Calories;
    Body_Temp -> Calories;
}
"""

# Initialize the causal model
model = CausalModel(
    data=df,
    treatment=['Duration', 'Heart_Rate', 'Body_Temp'],
    outcome='Calories',
    graph=graph
)

# Identify the causal effect
identified_estimand = model.identify_effect()

# Estimate the effect using linear regression
estimate = model.estimate_effect(
    identified_estimand,
    method_name="backdoor.linear_regression",
    control_value=0,
    treatment_value=1,
    confidence_intervals=True
)

print(estimate.value)  # Causal effect estimates
```

### Step 3: Build a Predictive Model
Using insights from DoWhy, we fit a regression model with key causal features (`Duration`, `Heart_Rate`, `Body_Temp`) and confounders (`Age`, `Sex`, `Height`, `Weight`).

```python
# Prepare data for prediction

y = df['Calories']

# Train a linear regression model

model_lr.fit(X, y)

# Predict calories for the first 5 entries
predictions = model_lr.predict(X.head())
print("Predicted Calories (first 5 samples):\n", predictions)
```
